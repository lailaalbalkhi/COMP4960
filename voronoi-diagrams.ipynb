```python
# creating a voronoi diagram example

import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt

coords = np.random.rand(10,2)
voronoi_diagram = Voronoi(coords)

fig = voronoi_plot_2d(voronoi_diagram)
plt.show()
```

```python
# load an image

import skimage as ski
from skimage import io, filters

img = io.imread('temp.jpg')

plt.imshow(img, cmap='gray')
plt.axis('off')
```

```python
# Step 1 - Gaussian Blur to average local densities

blurred_img = filters.gaussian(img, sigma=5)
plt.imshow(blurred_img, cmap='gray')
plt.axis('off')
```

```python
# Step 2 - Find points representing each object

from skimage.feature import peak_local_max

coordinates = peak_local_max(blurred_img, min_distance=20, exclude_border=False)
Vcoordinates = coordinates[:, :2]
print(Vcoordinates)

plt.imshow(blurred_img, cmap='gray')
plt.plot(Vcoordinates[:,1], Vcoordinates[:,0], 'r.')
plt.axis('off')
```

```python
# Step 3 - Finding Voronoi regions

voronoi = Voronoi(Vcoordinates)
print(Vcoordinates)
fig = voronoi_plot_2d(voronoi)

plt.axis('off')
plt.show()
```
